```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
netflix_data = pd.read_csv('/content/titles.csv')
```

```python
netflix_features = ["title", "description", "genres"]
print(netflix_features)
```

```python
for features in netflix_features:
  netflix_data[features] = netflix_data[features].fillna('')
```

```python
combining_netflix_features = netflix_data['title']+' '+netflix_data['description']+' '+netflix_data['genres']
print(combining_netflix_features)
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combining_netflix_features)
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
print(similarity.shape)
```

```python
netflix_name = input("Enter your movie name: ")
```

```python
list_of_all_netflix_title = netflix_data["title"].tolist()
print(list_of_all_netflix_title)
```

```python
finding_close_match = difflib.get_close_matches(netflix_name, list_of_all_netflix_title)
print(finding_close_match)
```

```python
close_match = finding_close_match[0]
print(close_match)
```

```python
index_of_the_netflix = netflix_data[netflix_data.title == close_match]['index'].values[0]
print(index_of_the_netflix)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_netflix]))
print(similarity_score)
len(similarity_score)
```

```python
sorted_similar_netflix = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_netflix)
```

```python
netflix_name = input("Enter your movie name: ")

list_of_all_netflix_title = netflix_data["title"].tolist()

finding_close_match = difflib.get_close_matches(netflix_name, list_of_all_netflix_title)

close_match = finding_close_match[0]

index_of_the_netflix = netflix_data[netflix_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_netflix]))

sorted_similar_netflix = sorted(similarity_score, key = lambda x:x[1], reverse = True) 


print("These are some reccomned shows \n")

i = 1 

for netflix in sorted_similar_netflix:
  index = netflix[0]
  title_from_index = netflix_data[netflix_data.index==index]['title'].values[0]
  if(i<31):
    print(i, ".", title_from_index)
    i+=1



```
